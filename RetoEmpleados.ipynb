9. Todo parece indicar que las variables categóricas que quedan sí son importantes para obtener la variable de salida. Convierte todas las variables categóricas que quedan a numéricas:

a. BusinessTravel

b. Department

c. EducationField

```python
# One-hot encoding

BusinessTravel_OH = pd.get_dummies(df_empleados1['BusinessTravel'], prefix = 'BusinessTravel').astype(int)
Department_OH = pd.get_dummies(df_empleados1['Department'], prefix = 'Department').astype(int)
EducationField_OH = pd.get_dummies(df_empleados1['EducationField'], prefix = 'EducationField').astype(int)
Gender_OH = pd.get_dummies(df_empleados1['Gender'], prefix = 'Gender').astype(int)
JobRole_OH = pd.get_dummies(df_empleados1['JobRole'], prefix = 'JobRole').astype(int)
MaritalStatus_OH = pd.get_dummies(df_empleados1['MaritalStatus'], prefix = 'MaritalStatus').astype(int)
Attrition_OH = pd.get_dummies(df_empleados1['Attrition'], prefix = 'Attrition').astype(int)
```

11. Selecciona solo aquellas variables que tengan una correlación mayor o igual a 0.1, dejándolas en otro frame llamado EmpleadosAttritionFinal. No olvides mantener la variable de salida Attrition; esto es equivalente a borrar las que no cumplen con el límite.


```python
#obtener correlación
df_empleados3 = df_empleados2.corr()

df_empleados3
```

```python
#creo mejor un un df donde solo venga la correlación con respecto a attrition
df_empleados3_attrition = df_empleados2.corr()['Attrition_Yes'].abs()

Cumplen = df_empleados3_attrition[df_empleados3_attrition >= 0.1].index

EmpleadosAttritionFinal = df_empleados2[Cumplen]
EmpleadosAttritionFinal.head()
```

12. Crea una nueva variable llamada EmpleadosAttritionPCA formada por los componentes principales del frame EmpleadosAttritionFinal. Recuerda que el resultado del proceso PCA es un numpy array, por lo que, para hacer referencia a una columna, por ejemplo, la 0, puedes usar la instrucción EmpleadosAttritionPCA[:,0]).


```python
#Como no se escalaron anteriormente las características puede que salga raro (solo la de ingreso)
from sklearn.decomposition import PCA

pca = PCA(n_components=EmpleadosAttritionFinal.shape[1])

EmpleadosAttritionPCA = pca.fit_transform(EmpleadosAttritionFinal)

print(EmpleadosAttritionPCA[:,0])
```

13. Agrega el mínimo número de Componentes Principales en columnas del frame EmpleadosAttritionPCA que logren explicar el 80% de la varianza, al frame EmpleadosAttritionFinal. Puedes usar la instrucción assign, columna por columna, llamando a cada una C0, C1, etc., hasta las que vayas a agregar.

```python
#Varianza acumulada de una mejor forma la pdf
var_acum = np.cumsum(pca.explained_variance_ratio_)

#Componentes que expliquen el 80%
componentes_80 = np.argmax(var_acum >= 0.8) + 1

print(componentes_80)
```

```python
# Necesitamos dos componentes
nuevas_columnas = {f'C{i}': EmpleadosAttritionPCA[:, i] for i in range(componentes_80)}

EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(**nuevas_columnas)

EmpleadosAttritionFinal.head()
```
